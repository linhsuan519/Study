#資料清理處理策略

Drop missing observations

Drop rows where all cells in that row is NA

Create a new column full of missing values

Fill in missing data with zeros

Fill in missing in preTestScore with the mean value of preTestScore

Fill in missing in postTestScore with each sex’s mean value of postTestScore

Select some rows but ignore the missing data points






```python
import numpy as np
import pandas as pd
raw_data = {'first_name': ['Jason', np.nan, 'Tina', 'Jake', 'Amy'],
        'last_name': ['Miller', np.nan, 'Ali', 'Milner', 'Cooze'],
        'age': [42, np.nan, 36, 24, 73],
        'sex': ['m', np.nan, 'f', 'm', 'f'],
        'preTestScore': [4, np.nan, np.nan, 2, 3],
        'postTestScore': [25, np.nan, np.nan, 62, 70]}
df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'sex', 'preTestScore', 'postTestScore'])
display(df)
```

```python
#Drop missing observations
df0 = df.copy()
df0.dropna(inplace = True)
display(df0)
print("_"*70)

#Drop rows where all cells in that row is NA
df1 = df.copy()
df1.dropna(thresh=4, inplace = True)#thresh保留至少4個非NaN
display(df1)
print("_"*70)

#Create a new column full of missing values
newcol = [np.nan, np.nan, np.nan, np.nan, np.nan]
df2 = df.copy()
df2['new'] = newcol
display(df2)
print("_"*70)

#Fill in missing data with zeros
df3 = df.copy()
df3.fillna(0, inplace = True)
display(df3)
print("_"*70)

#Fill in missing in preTestScore with the mean value of preTestScore
df4 = df.copy()
mean = df4.loc[0:4,'preTestScore'].mean()
print(mean)
df4.loc[0:4,'preTestScore'].fillna(mean, inplace = True)
display(df4)
print("_"*70)

#Fill in missing in postTestScore with each sex’s mean value of postTestScore
df5 = df.copy()
male = df5[(df5['sex'] == 'm') & (df5['postTestScore'].notna())]
mean_male = male['postTestScore'].mean()
male2 = df5[(df5['sex'] == 'm') & (df5['postTestScore'].isna())]
df5.loc[male2.index, 'postTestScore'] = mean_male
female = df5[(df5['sex'] == 'f') & (df5['postTestScore'].notna())]
mean_female = female['postTestScore'].mean()
female2 = df5[(df5['sex'] == 'f') & (df5['postTestScore'].isna())]
df5.loc[female2.index, 'postTestScore'] = mean_female
display(df5)
print("_"*70)

#Select some rows but ignore the missing data points
df6 = df.copy()
ignore = df6[df6['postTestScore'].notna()]
df6 = ignore.copy()
ignore2 = df6[df6['preTestScore'].notna()]
df6 = ignore2.copy()
display(df6)
```

```python
# Drop missing observations
display(df.dropna())

# Drop rows where all cells in that row is NA
display(df.dropna(how = 'all'))

# Create a new column full of missing values
Missing = pd.Series([np.nan] * df.shape[0], index = range(df.shape[0]), name = 'Missing')
display(pd.concat([df, Missing], axis = 1))

# Fill in missing data with zeros
display(df.fillna(0))

# Fill in missing in preTestScore with the mean value of preTestScore
display(df.fillna({'preTestScore': df.preTestScore.mean()}))

# Fill in missing in postTestScore with each sex’s mean value of postTestScore
df.postTestScore = df.groupby('sex').postTestScore.transform(lambda x: x.fillna(x.mean()))
display(df)

# Select some rows but ignore the missing data points
display(df.dropna().sample(n = 3))
```
